```python
# Importando as demais bibiotecas
from nltk.corpus import PlaintextCorpusReader
from nltk.corpus import stopwords
from matplotlib.colors import ListedColormap
from wordcloud import WordCloud
import string
```

```python
# importando o arquivo 
corpus = PlaintextCorpusReader('Arquivos', '.*')
corpus
```

```python
# Visualizando o arquivo 
arquivos = corpus.fileids()
arquivos[0]
arquivos[0:100]
for a in arquivos:
    print(a)
```

```python
# Vendo somente um texto. 
texto = corpus.raw('1.txt')
texto
```

```python
# Vendo todo o texto
todo_texto = corpus.raw()
palavras = corpus.words()
palavras[170]
len(palavras)
```

```python
# Criando a nuvem
stops = stopwords.words('english')
mapas_cores = ListedColormap(['orange','green','magenta'])
nuvem = WordCloud(background_color = 'white',
                  colormap = mapas_cores,
                  stopwords = stops,
                  max_words = 100)

nuvem.generate(todo_texto)
plt.imshow(nuvem)
```

```python
palavras_sem_stopwords = [ p for p in palavras if p not in stops]
len(palavras_sem_stopwords)
```

```python
# Pegando o conjunto de dados de palavras sem pontuação
palavras_sem_pontuacao = [p for p in palavras_sem_stopwords if p not in string.punctuation]
len(palavras_sem_pontuacao)
```

```python
# Calcular a frequencia das palafras sem pontuação
frequencia = nltk.FreqDist(palavras_sem_pontuacao)
mais_comuns = frequencia.most_common(100)
```
